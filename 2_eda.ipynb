```python
plt.figure(figsize= (8, 5))
sns.set_style('whitegrid')
sns.lineplot(x= df.year, y= df.white_median, marker= 'd', linewidth= 1.5, linestyle= "-.", color= "blue", markeredgecolor= 'yellow', markerfacecolor= 'green', 
             label= 'White Median')
sns.lineplot(x= df.year, y= df.black_median, marker= 'o', linewidth= 1.5, linestyle= '-.', color= 'red', markeredgecolor= 'yellow', markerfacecolor= 'purple', 
             label= 'Black Median')
sns.lineplot(x= df.year, y= df.white_average, marker= 'd', linewidth= 1.5, linestyle= "-.", color= "green", markeredgecolor= 'yellow', markerfacecolor= 'blue', 
             label= 'White Average')
sns.lineplot(x= df.year, y= df.black_average, marker= 'o', linewidth= 1.5, linestyle= '-.', color= 'purple', markeredgecolor= 'yellow', markerfacecolor= 'red', 
             label= 'Black Average')
plt.axhline(y= 0, linewidth= 1,linestyle= "-", color= 'black', label= 'Zero Wage')

plt.title('Black White Wage Comparison', fontsize= 16, fontweight= 'bold', color= 'black')
plt.xlabel('Year', fontsize= 14, fontweight= 'bold', color= 'slategray')
plt.ylabel('Value', fontsize= 14, fontweight= 'bold', color= 'slategray')

plt.grid(True, linewidth= 0.8, linestyle= '--', color= 'grey', alpha= 0.6)
plt.legend(title= 'Wage')
plt.tight_layout()
plt.show()
```
