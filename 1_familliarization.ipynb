```python
plt.style.use("seaborn-v0_8-bright")
plt.figure(figsize= (12,8))
plt.plot(df['year'], df['white_average'], color= "violet", linewidth= 3, linestyle= "--", marker= "o", label= 'White Average')
plt.plot(df['year'], df['black_average'], color= "indigo", linewidth= 3, linestyle= "-.", marker= "o", label= 'Black Average')
plt.plot(df['year'], df['white_median'], color= "blue", linewidth= 2, linestyle= ":", marker= "x", label= 'White Median')
plt.plot(df['year'], df['black_median'], color= "green", linewidth= 2, linestyle= "-", marker= "x", label= 'Black Median')
plt.title("Average and Median Wage over Time")
plt.xlabel("Year")
plt.ylabel("Wage")

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize= (12,8))
wage_gap= df['white_average'] - df['black_average']
wage_gap_median= df['white_median'] - df['black_median']

sns.lineplot(
    data=df,
    x='year',
    y=wage_gap,
    color='blue',
    linewidth=3,
    linestyle='-',
    marker='D',
    label='Wage Gap'
)
plt.title('Wage Gap Over Years')
plt.xlabel("Year")
plt.ylabel("Wage Gap")

plt.legend()
plt.show()
```

```python
plt.figure(figsize= (12, 8))
plt.plot(df['year'], wage_gap, label= "Wage Gap", linewidth= 3, marker= "o", linestyle= "--", color= "yellow", markerfacecolor= "yellow", markeredgecolor= "red")
plt.title("Wage Gap over Time")
plt.xlabel("Year")
plt.ylabel("Wage Gap")
plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8, label='Zero Gap')

plt.legend()
plt.grid(True, linestyle= "-.", color= "grey", alpha= 0.6)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize= (12, 8))
plt.plot(df['year'], wage_gap_median, label= "Wage Gap", linewidth= 3, marker= "o", linestyle= "--", color= "yellow", markerfacecolor= "yellow", markeredgecolor= "red")
plt.title("Wage Gap over Time")
plt.xlabel("Year")
plt.ylabel("Wage Gap")
plt.axhline(y= 0, color= 'black', linestyle= '--', linewidth= 0.8, label= 'Zero Gap')
[plt.axvline(x= x, color= 'black', linestyle= w, linewidth= 0.8, label= z) for w,x,z in [(":", 1979, 'Minimum'), ("-.", 2018, 'Maximum')]]
 
plt.legend()
plt.grid(True, linestyle= "-.", color= "grey", alpha= 0.6)
plt.tight_layout()
plt.show()
```

```python
gender_wage_diff_black= df['black_men_median'] - df['black_women_median']
gender_wage_diff_white= df['white_men_median'] - df['white_women_median']

plt.figure(figsize=(12, 8))
plt.plot(df['year'], gender_wage_diff_white, label= "Gender Wage Difference White", marker= "x", linewidth= 3, linestyle= '--', color= 'yellow',
        markeredgecolor= 'red', markerfacecolor= 'yellow')
plt.plot(df['year'], gender_wage_diff_black, label= "Gender Wage Difference Black", marker= "d", linewidth= 3, linestyle= "--", color= 'red',
        markeredgecolor= 'yellow', markerfacecolor= 'red')
plt.title("Gender Wage Difference Over Time")
plt.xlabel("Year")
plt.ylabel("Gender Wage Difference")
plt.axhline(y= 0, label= "Zero Difference", linewidth= 3, linestyle= "--")

plt.legend()
plt.tight_layout()
plt.grid(True, linewidth= 3, linestyle= '-.', color= "grey", alpha= 0.6)
plt.show()
```

```python
wage_difference_gender_men= df['white_men_median'] - df['black_men_median']
wage_difference_gender_women= df['white_women_median'] - df['black_women_median']

plt.figure(figsize= (12,8))
plt.plot(df['year'], wage_difference_gender_men, label= "Wage Difference Men", marker= "o", linewidth= 2, linestyle= '--', color= "purple", markeredgecolor= "yellow", markerfacecolor= "purple")
plt.plot(df['year'], wage_difference_gender_women, label= "Wage Difference Women", marker= "o", linewidth= 2, linestyle= '--', color= "green", markeredgecolor= "purple", markerfacecolor= "green")
plt.title("Wage Difference Based On Gender Over Time")
plt.xlabel("Year")
plt.ylabel("Wage Difference")
plt.axhline(y= 0, color= "black", linewidth= 3, linestyle= "--", label= "Zero Difference")

plt.legend()
plt.grid(True, color= "grey", linewidth= 2, linestyle= "-.", alpha= 0.6)
plt.tight_layout()
plt.show()
```
